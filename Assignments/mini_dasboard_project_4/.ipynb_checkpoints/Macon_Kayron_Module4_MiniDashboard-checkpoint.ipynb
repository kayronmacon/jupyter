```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (update path if needed)
df = pd.read_csv('module4_cyber_events-1.csv')
df.head()

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Visualization 1 – Matplotlib
Explain what this visualization shows.

```python
# Create a Matplotl# plot
fig = plt.figure(figsize=(18, 12))

ax1 = plt.subplot(2, 3, 1)
if 'event_type' in df.columns:
    event_counts = df['event_type'].value_counts()
elif 'category' in df.columns:
    event_counts = df['category'].value_counts()
elif 'type' in df.columns:
    event_counts = df['type'].value_counts()
else:
    # Use the first categorical column
    cat_cols = df.select_dtypes(include=['object']).columns
    if len(cat_cols) > 0:
        event_counts = df[cat_cols[0]].value_counts()
    else:
        event_counts = pd.Series([1])

colors = sns.color_palette("husl", len(event_counts))
event_counts.plot(kind='bar', ax=ax1, color=colors)
ax1.set_title('Distribution of Cyber Events by Type', fontsize=14, fontweight='bold')
ax1.set_xlabel('Event Type' )
ax1.set_ylabel('Number of Events')
ax1.tick_params(axis='x', rotation=45)
plt.setp(ax1.xaxis.get_majorticklabels(), ha='right')
```
