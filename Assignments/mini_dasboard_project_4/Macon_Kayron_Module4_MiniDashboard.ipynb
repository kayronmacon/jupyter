```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import textwrap
# Load the dataset (update path if needed)

df = pd.read_csv('module4_cyber_events-1.csv')
df.head()

```

## Visualization 1 – Matplotlib
Explain what this visualization shows.

```python
# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (24, 12)

# Create a Matplotl# plot
fig = plt.figure(figsize=(18, 12))

ax1 = plt.subplot(2, 3, 1)
event_counts = df['severity'].value_counts()
 # Use the first categorical column

colors = sns.color_palette("husl", len(event_counts))
event_counts.plot(kind='bar', ax=ax1, color=colors)
ax1.set_title('Distribution of Cyber Events Severity', fontsize=14, fontweight='bold')
ax1.set_xlabel('Severity Level')
ax1.set_ylabel('Number of Events')
ax1.tick_params(axis='x', rotation=45)

summary= f"""
This bar chart shows the overall frequency of events across different severity levels. It provides a high-level risk overview of the environment, highlighting whether the system is experiencing mostly low-level alerts or a significant number of high-severity incidents. This helps assess overall threat intensity."""
print(textwrap.fill(summary, width=80))
```

## Visualization 2 – Seaborn
Explain what this visualization shows.

```python
# Create a Seaborn visualization here

fig1, ax1 = plt.subplots(figsize=(12, 8))

# Simple scatter plot
ax1.scatter(df['severity'], df['source_ip'], 
            s=100,              # Size of points
            alpha=0.6,          # Transparency (0=transparent, 1=opaque)
            color='steelblue',  # Point color
            edgecolors='black', # Border around points
            linewidth=1)        # Border thickness

ax1.set_title('Severity vs Source IP', fontsize=16, fontweight='bold', pad=20)
ax1.set_xlabel('Severity Level', fontsize=13, fontweight='bold')
ax1.set_ylabel('Source IP', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

plt.tight_layout()
summary=f"""
This scatter plot shows the relationship between event severity and source IP addresses. It helps identify which IPs are generating high-severity events and whether threat activity is concentrated among specific sources. This visualization supports threat detection and prioritization by highlighting potentially malicious or high-risk IP addresses.
"""
print(textwrap.fill(summary, width=80))
```

## Visualization 3 – Your Choice
Explain what this visualization shows.

```python
# Convert timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])
plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=df,
    x='timestamp',
    y='severity',
    hue='event_type',
    size='severity',
    sizes=(50, 200),
    alpha=0.7
)

plt.title('Severity Levels by Timestamp', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Timestamp', fontsize=13, fontweight='bold')
plt.ylabel('Severity Level', fontsize=13, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(alpha=0.3, linestyle='--')

plt.tight_layout()
summary=f"""
This scatter plot displays cybersecurity event severity over time, with colors representing different event types. It helps identify trends, spikes, and periods of increased threat activity to support effective monitoring and response.
"""
print(textwrap.fill(summary, width=80))
```

## Dashboard Summary
Summarize the key insights from your visualizations.

```python
final_summary=f"""
These three visualizations provide a comprehensive overview of cybersecurity event activity. The severity distribution chart shows the overall threat level across the system, the severity versus source IP chart identifies where high-risk activity is originating, and the severity over time chart reveals when threats occur and highlights spikes or patterns in activity. Together, they offer a clear view of threat intensity, source concentration, and temporal trends to support effective monitoring and incident response.
"""
print(textwrap.fill(final_summary, width=80))
```
