```python
import pandas as pd

# Load dataset
df = pd.read_csv("mini_project_threat_logs.csv")

# Preview data
df.head()
```

```python
print("First 5 rows:")
print(df.head())

print("\nData types:")
print(df.dtypes)

print("\nSummary statistics:")
print(df.describe(include="all"))

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

print("\nDataset shape:", df.shape)
```

```python
# Create cleaned copy
df_cleaned = df.copy()

# Convert to datetime format
df_cleaned["start_time"] = pd.to_datetime(df_cleaned["start_time"], errors="coerce")
df_cleaned["end_time"] = pd.to_datetime(df_cleaned["end_time"], errors="coerce")

# Remove rows with missing critical values
df_cleaned = df_cleaned.dropna(subset=[
    "start_time",
    "end_time",
    "severity",
    "source_ip"
])

# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Reset index
df_cleaned = df_cleaned.reset_index(drop=True)

df_cleaned.head()
```
