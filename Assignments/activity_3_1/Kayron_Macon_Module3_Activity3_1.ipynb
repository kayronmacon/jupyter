**Instructions:** Follow the steps below to explore, filter, and clean a sample cybersecurity dataset using Pandas and NumPy.

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Load the dataset
df = pd.read_csv(r"C:\Users\kayro\jupyter\Assignments\activity_3_1\activity3_1_sample_dataset.csv")
df.head()
```

```python
# Step 4: Filter by alert_level > 5
high_alerts = df[df['alert_level'] > 5]
high_alerts
```

```python
# Step 5: Use NumPy to compute summary statistics
print("Mean packet size:", np.mean(df['packet_size']))
print("Max packet size:", np.max(df['packet_size']))
print("Min packet size:", np.min(df['packet_size']))
```

```python
# Step 6: Clean the data
df_clean = df.dropna(subset=['source_ip']).copy()
df_clean['alert_level'] = df_clean['alert_level'].fillna(0)
df_clean = df_clean.rename(columns={'packet_size': 'packet_bytes'})

# Step 7: Convert data type
df_clean['alert_level'] = df_clean['alert_level'].astype(int)
df_clean.head()
```
